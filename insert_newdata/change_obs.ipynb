```python
# 'AFR', 'AFR.EUR', 'AFR.SAS', 'EAS', 'EAS.EUR', 'EUR', 'EUR.SAS', 'SAS'

adata.obs['Ethnicity_inferred'].unique()

self_reported_ethnicity = list(adata.obs['Ethnicity_inferred'].values)

smoking_status = list(adata.obs['Smoker'].values)

condition = list(adata.obs['COVID_status'].values)

disease = list(adata.obs['COVID_status'].values)

for i in range(0,len(self_reported_ethnicity)):
    if condition[i] == 'Healthy':
        condition[i] = 'healthy'
        disease[i] = 'normal'
    if condition[i] == 'COVID+':
        condition[i] = 'COVID+'
        disease[i] = 'COVID+'
    
    
    if smoking_status[i] == 'Non-smoker':
        smoking_status[i] = 'never'
    if smoking_status[i] == 'Ex-smoker':
        smoking_status[i] = 'former'
    if smoking_status[i] == 'Smoker':
        smoking_status[i] = 'active'
    if smoking_status[i] == 'Unknown':
        smoking_status[i] = 'nan'
    
    if self_reported_ethnicity[i] in ['AFR', 'AFR.EUR', 'AFR.SAS']:
        self_reported_ethnicity[i] = 'African'
    if self_reported_ethnicity[i] in ['EAS', 'EAS.EUR','SAS']:
        self_reported_ethnicity[i] = 'Asian'
    if self_reported_ethnicity[i] in ['EUR', 'EUR.SAS']:
        self_reported_ethnicity[i] = 'European'


self_reported_ethnicity = pd.Series(self_reported_ethnicity,index=adata.obs_names)
adata.obs['self_reported_ethnicity'] = self_reported_ethnicity

smoking_status = pd.Series(smoking_status,index=adata.obs_names)
adata.obs['smoking_status'] = smoking_status

disease = pd.Series(disease,index=adata.obs_names)
adata.obs['disease'] = disease

condition = pd.Series(condition,index=adata.obs_names)
adata.obs['condition'] = condition
```
